## linear regression

using linear regression to explore relationship between building CO2 emission and energy consumption in La Defense

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12.0, 9.0)

#Preprocessing Input data
data = pd.read_csv('energyco2.csv',encoding="unicode_escape")
data.head()
```

```python
X = data.iloc[:, 7]
Y = data.iloc[:, 8]
plt.scatter(X,Y)
plt.show()
```

```python
# Building the model
A = 0
B = 0

L = 0.0001  # The learning Rate
epochs = 1000 

n = float(len(X)) # Number of elements in X

# Gradient Descent 
for i in range(epochs): 
    Y_pred = B*X + A  # The current predicted value of Y
    D_B = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt B
    D_A = (-2/n) * sum(Y - Y_pred)  # Derivative wrt A
    B = B - L * D_B  # Update B
    A = A - L * D_A  # Update A
print (A, B)
```

```python
# Making predictions
Y_pred = B*X + A

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
